```python
import pandas as pd

def actualizar_excel(excel1_path, excel2_path, output_path):
    # Leer los archivos de Excel especificando la hoja "FDE"
    df1 = pd.read_excel(excel1_path, sheet_name='FDE', skiprows=5)
    df2 = pd.read_excel(excel2_path, sheet_name='FDE', skiprows=5)

    # Crear una lista para almacenar los datos actualizados
    updated_rows = []

    # Iterar sobre las filas de df1
    for idx, row in df1.iterrows():
        ref_cliente = row.iloc[0]  # Ref. Cliente está en la columna A (índice 0)
        nombre_ref_cliente = row.iloc[1]  # Nombre Ref. Cliente está en la columna B (índice 1)
        cod_elemento = row.iloc[8]  # Cod elemento está en la columna I (índice 8)
        
        # Filtrar las filas correspondientes en df2
        matching_rows = df2[(df2.iloc[:, 0] == ref_cliente) &
                            (df2.iloc[:, 1] == nombre_ref_cliente) &
                            (df2.iloc[:, 8].str[:10] == str(cod_elemento)[:10])]
        
        if not matching_rows.empty:
            for _, match in matching_rows.iterrows():
                updated_rows.append(match)
        else:
            updated_rows.append(row)

    # Crear un nuevo DataFrame con las filas actualizadas
    df1_updated = pd.DataFrame(updated_rows, columns=df1.columns)

    # Guardar el archivo actualizado
    df1_updated.to_excel(output_path, index=False)

# Ejemplo de uso
excel1_path = r'C:\Users\soled\Desktop\Ecovidrio\files\ecovidrio\exceles\prueba.xlsm'
excel2_path = r'C:\Users\soled\Desktop\Ecovidrio\files\ecovidrio\exceles\FDE_Ventas_2023 MAHOU SA 300524 FINALNOMODIFICAR.xlsm'
output_path = r'C:\Users\soled\Desktop\Ecovidrio\files\ecovidrio\exceles\ruta_de_salida2.xlsx'
actualizar_excel(excel1_path, excel2_path, output_path)

```

```python
import pandas as pd

# Cargar los archivos de Excel
prueba_path = r'C:\Users\soled\Desktop\Ecovidrio\files\ecovidrio\eco\prueba.xlsm'
ecovidrio_path = r'C:\Users\soled\Desktop\Ecovidrio\files\ecovidrio\eco\MSM.xlsx'

# Cargar la hoja FDE desde prueba.xlsm y el nuevo archivo de precios desde ecovidrio.xlsx
prueba_df = pd.read_excel(prueba_path, sheet_name='FDE', skiprows=4)
ecovidrio_df = pd.read_excel(ecovidrio_path, sheet_name='Dato cer referencia', skiprows=4)

# Seleccionar las columnas relevantes de la hoja FDE
prueba_df = prueba_df[['Ref. Cliente', 'Nº de UD']]
prueba_df.columns = ['Codigo', 'Cantidad']

# Seleccionar las columnas relevantes de la hoja Dato cer referencia
ecovidrio_df = ecovidrio_df[['Ref. Cliente', 'NACIONAL', 'EXPORT']]
ecovidrio_df.columns = ['Codigo', 'Precio1', 'Precio2']

# Convertir las columnas a tipo numérico (si es necesario)
prueba_df['Cantidad'] = pd.to_numeric(prueba_df['Cantidad'], errors='coerce')
ecovidrio_df['Precio1'] = pd.to_numeric(ecovidrio_df['Precio1'], errors='coerce')
ecovidrio_df['Precio2'] = pd.to_numeric(ecovidrio_df['Precio2'], errors='coerce')

# Realizar el merge entre los dataframes
merged_df = pd.merge(prueba_df, ecovidrio_df, on='Codigo', how='left')

# Multiplicar las cantidades por los precios
merged_df['Valor_Precio1'] = merged_df['Cantidad'] * merged_df['Precio1']
merged_df['Valor_Precio2'] = merged_df['Cantidad'] * merged_df['Precio2']

# Guardar el resultado en un nuevo archivo de Excel
resultado_path = r'C:\Users\soled\Desktop\Ecovidrio\files\ecovidrio\eco\resultado1.xlsx'
merged_df.to_excel(resultado_path, index=False)

print("Operación completada. Archivo guardado en:", resultado_path)




```
